```python
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
data=pd.read_csv('/kaggle/input/cancer-data/Cancer_Data.csv')
print(data.columns)
print(data.describe())
print(data.info())
print(data.isna().sum())
print(data.diagnosis)
print(data.diagnosis.value_counts().values)
print('..............To check the data is balanced or not...........')
m=data.diagnosis.value_counts().values[0]/(data.diagnosis.value_counts().values[0]+data.diagnosis.value_counts().values[1])
b=data.diagnosis.value_counts().values[1]/(data.diagnosis.value_counts().values[0]+data.diagnosis.value_counts().values[1])
if m >70.000 or b <30.000:
    print('The data is balanced')
num_col=data.select_dtypes(include='number').columns.values
print(num_col)


data['z-scores']=(data.radius_mean-data.radius_mean.mean())/(data.radius_mean.std())
df=data[(data['z-scores'] >-3)&(data['z-scores']<3)]
q1=df.radius_mean.quantile(0.25)
q3=df.radius_mean.quantile(0.75)
iqr=q3-q1
up=q3+1.5*iqr
lo=q1-1.5*iqr
df=df[(df.radius_mean <up)&(df.radius_mean >lo)]

q1u=df.radius_mean.quantile(0.25)
q3u=df.radius_mean.quantile(0.75)
i_qr=q3u-q1u
u_p=q3u+1.5*i_qr
l_o=q1u-1.5*i_qr
df=df[(df.radius_mean <u_p)&(df.radius_mean >l_o)]

qu1=df.compactness_mean.quantile(0.25)
qu3=df.compactness_mean.quantile(0.75)
iqr_=qu3-qu1
uppe=qu3+1.5*iqr_
lowe=qu1-1.5*iqr_
df=df[(df.compactness_mean <uppe)&(df.compactness_mean >lowe)]

qua1=df.texture_mean.quantile(0.25)
qua3=df.texture_mean.quantile(0.75)
iq_r=qua3-qua1
upper=qua3+1.5*iq_r
lower=qua1-1.5*iq_r
df=df[(df.texture_mean <upper)&(df.texture_mean >lower)]

quan1=df.perimeter_mean.quantile(0.25)
quan3=df.perimeter_mean.quantile(0.75)
i_q_r=quan3-quan1
uppe_r=quan3+1.5*i_q_r
lowe_r=quan1-1.5*i_q_r
df=df[(df.perimeter_mean <uppe_r)&(df.perimeter_mean >lowe_r)]

quant1=df.area_mean.quantile(0.25)
quant3=df.area_mean.quantile(0.75)
i_q_r=quant3-quant1
upp_er=quant3+1.5*i_q_r
low_er=quant1-1.5*i_q_r
df=df[(df.area_mean <upp_er)&(df.area_mean >low_er)]

quanti1=df.area_mean.quantile(0.25)
quanti3=df.area_mean.quantile(0.75)
Iq_r=quanti3-quanti1
up_p_er=quanti3+1.5*Iq_r
lo_w_er=quanti1-1.5*Iq_r
df=df[(df.area_mean <up_p_er)&(df.area_mean >lo_w_er)]

quantile1=df.area_mean.quantile(0.25)
quantile3=df.area_mean.quantile(0.75)
IQ_r=quantile3-quantile1
u_p_p_er=quantile3+1.5*IQ_r
l_o_w_er=quantile1-1.5*IQ_r
df=df[(df.area_mean <u_p_p_er)&(df.area_mean >l_o_w_er)]


quantiles1=df.concavity_mean.quantile(0.25)
quantiles3=df.concavity_mean.quantile(0.75)
IQ_R=quantiles3-quantiles1
u_p_p_e_r=quantiles3+1.5*IQ_R
l_o_w_e_r=quantiles1-1.5*IQ_R
df=df[(df.concavity_mean <u_p_p_e_r)&(df.concavity_mean >l_o_w_e_r)]



quantil11=df.concavity_se.quantile(0.25)
quantil33=df.concavity_se.quantile(0.75)
IQ_=quantil33-quantil11
u_p_per=quantil33+1.5*IQ_
l_o_wer=quantil11-1.5*IQ_
df=df[(df.concavity_se <u_p_per)&(df.concavity_se >l_o_wer)]

quantile11=df.compactness_worst.quantile(0.25)
quantile33=df.compactness_worst.quantile(0.75)
IQ_R=quantile33-quantile11
u_ppe_r=quantile33+1.5*IQ_R
l_owe_r=quantile11-1.5*IQ_R
df=df[(df.compactness_worst <u_ppe_r)&(df.compactness_worst >l_owe_r)]

quantil11=df.compactness_se.quantile(0.25)
quantil33=df.compactness_se.quantile(0.75)
I_Q_R=quantil33-quantil11
u_pper=quantil33+1.5*I_Q_R
l_ower=quantil11-1.5*I_Q_R
df=df[(df.compactness_se <u_pper)&(df.compactness_se >l_ower)]
print(data.shape)
print(df.shape)


x=df[['radius_mean' ,'texture_mean' ,'perimeter_mean'
 ,'smoothness_mean' ,'compactness_mean'
 ,'fractal_dimension_mean'
 ,'compactness_se','concave points_se'
 ,'texture_worst'
 ,'smoothness_worst' ,'compactness_worst' ,'concavity_worst']]
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['diagnose']=lab.fit_transform(df['diagnosis'])
y=df['diagnose']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
print('Logistic regression->: ',lr.score(x_test,y_test))

tre=DecisionTreeClassifier()
tre.fit(x_train,y_train)
print('Decision tree ',tre.score(x_test,y_test))

xgb=XGBClassifier()
xgb.fit(x_train,y_train)
print('XGB ',xgb.score(x_test,y_test))


rf=RandomForestClassifier()
rf.fit(x_train,y_train)
print('Random forest ',rf.score(x_test,y_test))

lgb=LGBMClassifier()
lgb.fit(x_train,y_train)
print('Light GBM ',lgb.score(x_test,y_test))
```
